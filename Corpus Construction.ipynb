```python
def addDF(urls, platform, game, team, bigdf):
  inc = 0
  column = 'message' if platform == 'Facebook' else 'snippet.topLevelComment.snippet.textOriginal' # different column names based on platform extracted from
  separator = ';' if platform == 'Facebook' else ','
  for url in urls:
    temp_df = pd.read_csv(url, sep=separator)
    for ind in temp_df.index:
      bigdf.loc[len(bigdf.index)] = [platform, game[inc], team[inc], temp_df.loc[ind][column]]
    inc = inc + 1
```

# Corpus Cleaning

```python
# checkpoint
fulldf = pd.read_csv('https://raw.githubusercontent.com/javi-ng/datasets/main/CSCI%20161%20Project%20Datasets/full_df.csv')
```

```python
fulldf
```

```python
# Instatiate NLTK module
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('stopwords')
stopwords = stopwords.words('english')
lemmatizer = WordNetLemmatizer()
```

```python
def clean_comments(comments):
    # Transform into lowercase
    comments['cleaned_text'] = comments['Entry'].str.lower()

    # Remove usernames, non-alphanumeric characters, and links
    comments['cleaned_text'] = comments['cleaned_text'].str.replace(r'(@[A-Za-z0-9_]+)|([^A-Za-z0-9_ \t])|(\w+:\/\/\S+)', '')

    # Remove single characters
    comments['cleaned_text'] = comments['cleaned_text'].apply(lambda row: ' '.join([word for word in row.split() if len(word) > 1]))

    # Remove stopwords
    comments['cleaned_text'] = comments['cleaned_text'].apply(lambda row: ' '.join([word for word in row.split() if word not in (stopwords)]))

    # Lemmatize verbs
    comments['cleaned_text'] = comments['cleaned_text'].apply(lambda row: ' '.join([lemmatizer.lemmatize(x, 'v') for x in row.split()]))

    # Lemmatize adjectives
    comments['cleaned_text'] = comments['cleaned_text'].apply(lambda row: ' '.join([lemmatizer.lemmatize(x, 'a') for x in row.split()]))

    # Lemmatize nouns
    comments['cleaned_text'] = comments['cleaned_text'].apply(lambda row: ' '.join([lemmatizer.lemmatize(x, 'n') for x in row.split()]))

    # Remove trailing and leading whitespaces
    comments['cleaned_text'] = comments['cleaned_text'].str.replace(r'^\s+|\s+$', '')

    return comments
```

```python
clean_df = clean_comments(fulldf)
clean_df
```

```python
clean_df = clean_df.drop(labels="Entry", axis=1)
```

```python
clean_df
```

```python
clean_df = clean_df.replace(r'^\s*$', np.nan, regex=True)
clean_df
```

```python
clean_df = clean_df.dropna(how='any', axis=0).reset_index(drop=True)
clean_df
```

```python
clean_df.to_csv('clean_df.csv', index=False)
```
